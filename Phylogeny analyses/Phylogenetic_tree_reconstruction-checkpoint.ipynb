```python
import os
import subprocess

# Define the directories containing the sequence files
seq_directory = 'All seq'  # Replace with your path
# Define the output directories for aligned files
aligned_seq_directory = 'aligned'  # Replace with your path

mafft_thread = 20

def align_sequences(file_path, output_dir):
    """Align sequences in a file using MAFFT and save to the specified output directory."""
    file_name = os.path.basename(file_path)
    base_name, ext = os.path.splitext(file_name)
    # Ensure the output file ends with '_aligned.fasta'
    output_file_name = f"{base_name}_aligned.fasta"
    output_path = os.path.join(output_dir, output_file_name)
    
    mafft_command = ['mafft', '--thread', str(mafft_thread), '--auto', file_path]
    with open(output_path, 'w') as output_file:
        subprocess.run(mafft_command, stdout=output_file)

def process_directory(input_dir, output_dir):
    """Process all sequence files in a given input directory and store aligned files in the output directory."""
    # Create the output directory if it does not exist
    os.makedirs(output_dir, exist_ok=True)
    
    # Get a list of all files that will be processed
    valid_extensions = ('.fasta', '.fas', '.faa', '.fna')
    files_to_process = [f for f in os.listdir(input_dir) if f.endswith(valid_extensions)]
    total_files = len(files_to_process)
    
    for index, file_name in enumerate(files_to_process, start=1):
        file_path = os.path.join(input_dir, file_name)
        align_sequences(file_path, output_dir)
        progress_percentage = (index / total_files) * 100
        print(f"Alignment complete for file: {file_name}. Progress: {progress_percentage:.2f}%")

# Perform alignment for the sequence directory
process_directory(seq_directory, aligned_seq_directory)

print("All alignments are complete.")
```

## 2. Trimming 

Clipkit

```python
import os
import subprocess

# Define the directories containing the aligned .fa files
aligned_seq_directory = 'aligned'  # Replace with your path

# Define the output directories for trimmed files
trimmed_seq_directory = 'trimmed'  # Replace with your path

def trim_sequences(file_path, output_dir):
    """Trim sequences in a file using ClipKIT and save to the specified output directory."""
    file_name = os.path.basename(file_path)
    output_path = os.path.join(output_dir, file_name.replace('_aligned.fasta', '_trimmed.fasta'))
    
#    clipkit_command = ['clipkit', file_path, '-m', 'gappy', '-g' '1','-o', output_path]
    clipkit_command = ['clipkit', file_path, '-m', 'smart-gap', '-o', output_path]
    subprocess.run(clipkit_command)

def process_directory(input_dir, output_dir):
    """Process all aligned files in a given input directory and store trimmed files in the output directory."""
    # Create the output directory if it does not exist
    os.makedirs(output_dir, exist_ok=True)

    for file_name in os.listdir(input_dir):
        if '_aligned.fasta' in file_name:
            file_path = os.path.join(input_dir, file_name)
            trim_sequences(file_path, output_dir)
            print(f"Trimming complete for file: {file_name}")

# Perform trimming for the .fa directory
process_directory(aligned_seq_directory, trimmed_seq_directory)

print("All trimming operations for .fa files are complete.")
```

## 3. Modeltest

modeltest-ng

```python
import os
import re
import subprocess
from Bio import AlignIO
from pathlib import Path

# Set up base directory and subdirectories
base_dir = os.getcwd()
trimmed_seq_folder = os.path.join(base_dir, "trimmed")
modeltest_seq_folder = os.path.join(base_dir, "modeltest_seqs")

# Create the modeltest_fa directory if it doesn't exist
os.makedirs(modeltest_seq_folder, exist_ok=True)

def mb_model(model, index):
    return f"lset applyto=(part{index}) model={model};\n"

# List all the clipped files in the directory
clips = list(Path(trimmed_seq_folder).glob("*.fasta"))

start = 1
partition = ""
mb_charset = ""
mb_lset = ""

for index, clip in enumerate(clips):
    msa = AlignIO.read(str(clip), "fasta")
    msa_len = len(msa[0].seq)

    # Model test for iqtree and RAxML
    modeltest_path = os.path.join(modeltest_seq_folder, str(index))
    subprocess.run(["modeltest-ng", "-i", str(clip), "-d", "nt", "-o", modeltest_path])
    
    try:
        with open(modeltest_path + ".out", 'r') as f:
            out = f.read()
        regexp = re.search("raxml-ng --msa.*--model (.*)", out)
        model = regexp.group(1)
        partition += f"{model}, gene{index}={start}-{start+msa_len-1}\n"
    except Exception as e:
        print(f"Error processing {clip}: {e}")

    # Model test for MrBayes
    modeltest_path_mb = os.path.join(modeltest_seq_folder, "mb" + str(index))
    subprocess.run(["modeltest-ng", "-i", str(clip), "-d", "nt", "-o", modeltest_path_mb, "-T", "mrbayes"])

    try:
        with open(modeltest_path_mb + ".out", 'r') as f:
            out = f.read()
        regexp = re.search("Model:              (.*)", out)
        model = regexp.group(1)
        mb_charset += f"charset part{index}={start}-{start+msa_len-1};\n"
        mb_lset += mb_model(model, index+1)
    except Exception as e:
        print(f"Error processing MrBayes model for {clip}: {e}")

    start += msa_len

mb_charset += f"partition favored = {len(clips)}: {', '.join(['part'+str(i) for i in range(len(clips))])};\n"
mb_charset += "set partition = favored;\n"

# Save files
Path(os.path.join(modeltest_seq_folder, 'part.txt')).write_text(partition)
Path(os.path.join(modeltest_seq_folder, 'mb_lset.txt')).write_text(mb_lset)
Path(os.path.join(modeltest_seq_folder, 'mb_charset.txt')).write_text(mb_charset)
```

## 4. Concatnate sequences

```python
import os
from Bio import SeqIO

# Define the directory containing the FASTA files
directory = "trimmed"

# Collect all sequences from the files
all_sequences = {}
lengths = {}

# Loop through all files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".fasta"):
        file_path = os.path.join(directory, filename)
        sequences = {record.id: str(record.seq) for record in SeqIO.parse(file_path, "fasta")}
        for seq_id, sequence in sequences.items():
            if seq_id not in all_sequences:
                all_sequences[seq_id] = {}
            all_sequences[seq_id][filename] = sequence
            if filename not in lengths:
                lengths[filename] = len(sequence)
            else:
                lengths[filename] = max(lengths[filename], len(sequence))

# Get the set of all sequence IDs
all_titles = set(all_sequences.keys())

# Concatenate sequences and fill missing sequences with '?'
concatenated_sequences = {}
for title in all_titles:
    concatenated_sequence = ""
    for filename in os.listdir(directory):
        if filename.endswith(".fasta"):
            seq = all_sequences[title].get(filename, '?' * lengths[filename])
            concatenated_sequence += seq
    concatenated_sequences[title] = concatenated_sequence

# Convert the result to a FASTA formatted string
concatenated_fasta = ""
for title, sequence in concatenated_sequences.items():
    concatenated_fasta += f">{title}\n{sequence}\n"

# Save the concatenated sequences to a new FASTA file
output_file = "concatenated.fasta"
with open(output_file, 'w') as file:
    file.write(concatenated_fasta)

output_file
```

### 5.2 MrBayes

Bayesian inference

```python
# Function to translate model settings for MrBayes
def mb_model(model, applyto):
    r = ""
    MRBAYES_AVAILABLE_MODEL = {
        "GTR": {
            "nst": "6",
        },
        "SYM": {
            "nst": "6",
            "statefreqpr": "fixed(equal)"
        },
        "HKY": {
            "nst": "2",
        },
        "K80": {
            "nst": "2",
            "statefreqpr": "fixed(equal)"
        },
        "F81": {
            "nst": "1",
        },
        "JC": {
            "nst": "1",
            "statefreqpr": "fixed(equal)"
        },
    }
    rates = "equal"
    if "+I+G4" in model:
        rates = "invgamma"
        model = model.replace("+I+G4", "")
    if "+I" in model:
        rates = "propinv"
        model = model.replace("+I", "")
    if "+G4" in model:
        rates = "gamma"
        model = model.replace("+G4", "")
    model_conf = MRBAYES_AVAILABLE_MODEL[model]
    r += "lset applyto=({}) nst={} rates={};\n".format(applyto, model_conf["nst"], rates)
    if "statefreqpr" in model_conf:
        r += "prset applyto=({}) statefreqpr={};\n".format(applyto, model_conf["statefreqpr"])
    return r
```

```python
import os

# Set the input file paths
fasta_file_path = 'concatenated.fasta'
charset_file_path = 'modeltest_seqs/mb_charset.txt'
lset_file_path = 'modeltest_seqs/mb_lset.txt'
nexus_file_path = 'mrbayes/mb.nex'

# Ensure the output directory exists
output_directory = os.path.dirname(nexus_file_path)
os.makedirs(output_directory, exist_ok=True)

# Read sequences from concatenated.fasta
with open(fasta_file_path, 'r') as fasta_file:
    sequences = fasta_file.read()

# Count the number of taxa and characters
num_taxa = sequences.count('>')

# Extract sequences and their lengths
sequence_list = sequences.split('>')
sequence_list = [seq for seq in sequence_list if seq]  # Remove empty entries
sequences_dict = {}
for seq in sequence_list:
    lines = seq.split('\n')
    taxon = lines[0]
    sequence = ''.join(lines[1:])
    sequences_dict[taxon] = sequence

num_characters = len(next(iter(sequences_dict.values())))

# Read charset information from mb_charset.txt
with open(charset_file_path, 'r') as charset_file:
    charset_info = charset_file.read()

# Read model settings from mb_lset.txt
with open(lset_file_path, 'r') as lset_file:
    lset_info = lset_file.read()

# Extract partition information from charset_info
partition_lines = charset_info.split('\n')
charset_lines = [line for line in partition_lines if line.startswith('charset')]
partition_definition = [line for line in partition_lines if line.startswith('partition')][0]

# Define the content of the NEXUS file
nexus_content = f"""#NEXUS

BEGIN DATA;
    DIMENSIONS NTAX={num_taxa} NCHAR={num_characters};
    FORMAT DATATYPE=DNA MISSING=? GAP=-;
    MATRIX
"""

for taxon, sequence in sequences_dict.items():
    nexus_content += f"    {taxon}    {sequence}\n"

nexus_content += """    ;
END;

BEGIN SETS;
"""

for line in charset_lines:
    nexus_content += f"    {line}\n"

nexus_content += """END;

BEGIN PARTITIONS;
"""

nexus_content += f"    {partition_definition}\n"

nexus_content += """END;

BEGIN MRBAYES;
    set seed=5566;
    set swapseed=5566;
"""

for line in charset_lines:
    nexus_content += f"    {line}\n"

nexus_content += f"""
    {partition_definition}
    set partition=favored;
"""

# Append model settings using mb_model function
lset_lines = lset_info.split('\n')
part_num = 1
for line in lset_lines:
    if line.strip():
        parts = line.split()
        model = parts[2].split('=')[1].strip(';')
        nexus_content += mb_model(model, part_num)
        part_num += 1

nexus_content += """
    mcmcp ngen=1000000 nruns=2 samplefreq=100 savebrlens=yes nchains=4 filename=mb;
    mcmc;
    sumt filename=mb burnin=2000 contype=allcompat showtreeprobs=no;
END;
"""

# Write the content to input.nex
with open(nexus_file_path, 'w') as nex_file:
    nex_file.write(nexus_content)

print(f'nexus saved to {nexus_file_path}')
```
